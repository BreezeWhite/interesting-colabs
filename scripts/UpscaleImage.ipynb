```python
#@title Upload Image

from google.colab import files
import IPython.display as dsp

ff = files.upload()
img_name = list(ff.keys())[0]
dsp.display(dsp.Image(img_name))
```

```python
#@title Run

import numpy as np
from PIL import Image
from ISR.models import RDN, RRDN

img = np.array(Image.open(img_name))

weights = 'psnr-small'  #@param ['psnr-small', 'psnr-large', 'noise-cancel', 'gans']
model = RRDN(weights=weights) if weights=='gans' else RDN(weights=weights)

sr_img = model.predict(np.array(img))
dsp.display(Image.fromarray(sr_img))
```
