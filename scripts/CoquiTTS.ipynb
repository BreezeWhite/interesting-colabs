```python
#@title Run

from IPython.display import display, Audio, HTML, clear_output


text = "A woman is yelling madly and loudly at a innocent cat in front of a table during the dinner in a France restaurant."  #@param {type:"string"}
tts_model = 'tacotron2-DCA'  #@param ['glow-tts', 'speedy-speech', 'tacotron2', 'tacotron2-DCA', 'tacotron2-DDC', 'tacotron2-DDC_ph', 'tacotron-DDC_sa', 'vits-lj', 'vits-vc', 'fast_pitch-lj', 'fast_pitch-vc']
vocoder = 'multiband-melgan'  #@param ['fullband-melgan', 'multiband-melgan', 'univnet', 'wavegrad-ek', 'wavegrad-li', 'hifigan_v2-lj', 'hifigan_v2-vc', 'hifigan_v2-sa']

#@markdown ### Special for Japanese input (yes, I'm much biased ^^)
use_jp = False  #@param {type:"boolean"}


tts_model = {
  'glow-tts': 'tts_models/en/ljspeech/glow-tts',
  'speedy-speech': 'tts_models/en/ljspeech/speedy-speech',
  'tacotron2': 'tts_models/en/ek1/tacotron2',
  'tacotron2-DCA': 'tts_models/en/ljspeech/tacotron2-DCA',
  'tacotron2-DDC': 'tts_models/en/ljspeech/tacotron2-DDC',
  'tacotron2-DDC_ph': 'tts_models/en/ljspeech/tacotron2-DDC_ph',
  'tacotron-DDC_sa': 'tts_models/en/sam/tacotron-DDC',
  'vits-lj': 'tts_models/en/ljspeech/vits',
  'vits-vc': 'tts_models/en/vctk/vits',
  'fast_pitch-lj': 'tts_models/en/ljspeech/fast_pitch',
  'fast_pitch-vc': 'tts_models/en/vctk/fast_pitch',
}[tts_model]

vocoder = {
  'fullband-melgan': 'vocoder_models/universal/libri-tts/fullband-melgan',
  'multiband-melgan': 'vocoder_models/en/ljspeech/multiband-melgan',
  'univnet': 'vocoder_models/en/ljspeech/univnet',
  'wavegrad-ek': 'vocoder_models/en/ek1/wavegrad',
  'wavegrad-li': 'vocoder_models/universal/libri-tts/wavegrad',
  'hifigan_v2-lj': 'vocoder_models/en/ljspeech/hifigan_v2',
  'hifigan_v2-vc': 'vocoder_models/en/vctk/hifigan_v2',
  'hifigan_v2-sa': 'vocoder_models/en/sam/hifigan_v2',
}[vocoder]

if use_jp:
  tts_model = 'tts_models/ja/kokoro/tacotron2-DDC'
  vocoder = 'vocoder_models/ja/kokoro/hifigan_v1'

exp_jp = ""

!tts \
  --use_cuda True \
  --text "{text}" \
  --model_name {tts_model} \
  --vocoder_name {vocoder} \
  --out_path ./out.wav

clear_output()
display(HTML(f'<h2>{text}</h2>'))
display(Audio('out.wav'))
```
