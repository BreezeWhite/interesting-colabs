```python
#@title Upload Audio
#@markdown **Please upload audio clip less than 30 seconds, or the memory comsumption would be unaffordable.**

import os
from google.colab import files
from IPython.display import display, Audio, HTML


ff = files.upload()
audio_name = list(ff.keys())[0]
out_dir = os.path.splitext(audio_name)[0]
os.makedirs(out_dir, exist_ok=True)
display(Audio(audio_name))
```

```python
#@title Run

!danna_sep {audio_name} --outdir {out_dir}

for track in ['vocals', 'drums', 'bass', 'other']:
  track_name = f'{out_dir}/{out_dir}_{track}'
  !ffmpeg -loglevel quiet -i {track_name}.wav {track_name}.mp3 

display(HTML('<h2>Vocals</h2>'))
display(Audio(f'{out_dir}/{out_dir}_vocals.mp3'))
display(HTML('<h2>Drums</h2>'))
display(Audio(f'{out_dir}/{out_dir}_drums.mp3'))
display(HTML('<h2>Bass</h2>'))
display(Audio(f'{out_dir}/{out_dir}_bass.mp3'))
display(HTML('<h2>Other</h2>'))
display(Audio(f'{out_dir}/{out_dir}_other.mp3'))
```
