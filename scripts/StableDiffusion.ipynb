```python
#@title Setup

# Solution to the cuda issue: https://github.com/tensorflow/models/issues/10590#issuecomment-1104552867
!apt install --allow-change-held-packages libcudnn8=8.1.0.77-1+cuda11.2
!pip install --upgrade keras-cv tensorflow

%load_ext autoreload
%autoreload 2

import keras_cv
from tensorflow import keras

keras.mixed_precision.set_global_policy("mixed_float16")
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512, jit_compile=True)
```

```python
#@title Run

from PIL import Image
from IPython.display import display

import numpy as np
import keras_cv
from tensorflow import keras

text = "made in abyss, dangerous valley, intricate, cinematic, glowing lights, hyperdetailed, highly detailed" #@param {type:"string"}
num_images = 3 #@param {type:"integer"}

keras.backend.clear_session()
images = model.text_to_image(text, batch_size=num_images)

out = np.concatenate(images, axis=1)
display(Image.fromarray(out))
```
